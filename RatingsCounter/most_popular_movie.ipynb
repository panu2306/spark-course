```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster('local').setAppName('MostPopularMovies')
```

```python
sc = SparkContext(conf=conf)
```

```python
input = sc.textFile('ml-100k/u.data')
```

```python
input.collect()
```

```python
movies = input.map(lambda x: (int(x.split()[1]), 1))
```

```python
movies.collect()
```

```python
moviesCounts = movies.reduceByKey(lambda x,y: x+y)
```

```python
moviesCounts.collect()
```

```python
flipped = moviesCounts.map(lambda x: (x[1], x[0]))
```

```python
flipped.collect()
```

```python
sortedMovies = flipped.sortByKey()
```

```python
results = sortedMovies.collect()
```

```python
for result in results: 
    print(result)
```
