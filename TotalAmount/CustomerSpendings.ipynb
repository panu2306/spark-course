```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster('local').setAppName('CustomerSpents')
```

```python
sc = SparkContext(conf=conf)
```

```python
input = sc.textFile('customer-orders.csv')
```

```python
def parsed_line(line):
    fields = line.split(',')
    cust_id = int(fields[0])
    spending = float(fields[2])
    return (cust_id, spending)
```

```python
lines = input.map(parsed_line)
```

```python
lines.collect()
```

```python
customer_spendings = lines.reduceByKey(lambda x, y: x + y)
```

```python
customer_spendings.collect()
```

```python
customer_spendings = customer_spendings.sortByKey()
```

```python
customer_spendings.collect()
```
